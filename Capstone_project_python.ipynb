```python
### Data visualization
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
##Scatter plot
plt.figure(figsize = (14,10))
plt.scatter(df["Acquisition_Cost"],df["ROI"],alpha = 0.6)
plt.xlabel("Acquisition_Cost")
plt.ylabel("ROI")
plt.title("Acquisition_Cost V/S ROI ")
plt.grid(True)
plt.show()
```

```python
###Bar plot
grouped =df.groupby(["Channel_Used","Campaign_Type"])["Conversion_Rate"].mean().reset_index()
plt.figure(figsize = (8,5))
sns.barplot(data = grouped,x="Channel_Used",y = "Conversion_Rate",hue="Campaign_Type")
plt.title("Average conversion rate by channel used,campaign type")
plt.ylabel("Average conversion rate")
plt.xticks(rotation = 45)
plt.show()

```

```python
### Box plot
plt.figure(figsize=(8,5))
sns.boxplot(data=df,x="Campaign_Type",y="Engagement_Score")
plt.title("Engagement score v/s campaign type")
plt.xlabel("Campaign_Type")
plt.ylabel("Engagement_Score")
plt.xticks(rotation = 45)
plt.show()
```

```python
##Bar plot
avg_roi = df.groupby("Company")["ROI"].mean().reset_index()
plt.figure(figsize = (8,5))
sns.barplot(data = avg_roi,x ="Company",y="ROI",palette = "viridis")
plt.title("Average ROI by company")
plt.xlabel("company")
plt.ylabel("Average ROI")
plt.xticks(rotation = 45)
plt.show()
```

```python
###Heat map
correlation_data = df[["Engagement_Score","Conversion_Rate"]]
correlation_matrix = correlation_data.corr()
plt.figure(figsize = (8,5))
sns.heatmap(correlation_matrix,annot = True,cmap = "coolwarm",fmt=".2f")
plt.title("correlation between engagement score and conversion rate")
plt.show()
```

```python
###Count plot
plt.figure(figsize = (8,5))
sns.countplot(data = df, x="Target_Audience",palette="Set2")
plt.title("Distribution of Target audience")
plt.xlabel("Target_Audience")
plt.ylabel("count")
plt.xticks(rotation = 45)
plt.show()
```

```python
###Bar plot
max_conversion = df.loc[df.groupby("Language")["Conversion_Rate"].idxmax()]
plt.figure(figsize = (8,5))
sns.barplot(data=max_conversion,x="Language",y="Conversion_Rate",hue ="Customer_Segment",dodge=False )
plt.title("customer segment for highest conversion rate per language")
plt.xlabel("Language")
plt.ylabel("Conversion_Rate")
plt.legend(title="Customer_Segment")
plt.show()
```

```python
###Box plot
plt.figure(figsize = (8,5))
sns.boxplot(x ="Customer_Segment",y="Acquisition_Cost",hue="Channel_Used",data=df)
plt.title(" Distribution of Acquisition Cost by Customer Segment,Channel Used")
plt.xlabel("Customer_Segment")
plt.ylabel("Acquisition_Cost")
plt.legend(title = "Channel Used")
plt.show()
```

```python
### Bar plot
avg_conversion_language = df.groupby("Language")["Conversion_Rate"].mean().reset_index()
avg_conversion_language = avg_conversion_language.sort_values(by ="Conversion_Rate",ascending=False)
print("avg_conversion_language")
plt.figure(figsize=(8,5))
plt.bar(avg_conversion_language["Language"],avg_conversion_language["Conversion_Rate"],color ="green")
plt.title("Average conversion rate by language")
plt.xlabel("Language")
plt.ylabel("Average conversion rate")
plt.xticks(rotation = 45)
plt.tight_layout()
plt.show()
```

```python
#### Chanel effectivness
### bar chart
Avg_engagement = df.groupby(["Campaign_Type","Channel_Used"])["Engagement_Score"].mean().reset_index()
print("Avg_engagement")
plt.figure(figsize=(8,5))
sns.barplot(data=Avg_engagement,x="Channel_Used",y="Engagement_Score",hue="Campaign_Type")
plt.title("Avg engagement by campaigntype and channel used")
plt.xlabel("Campaign_Type")
plt.ylabel("Avg_engagement_score")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
####Pie chart
total_roi_by_channel = df.groupby("Channel_Used")["ROI"].sum()
print("total_roi_by_channel")
plt.figure(figsize=(8,5))
plt.pie(total_roi_by_channel,labels=total_roi_by_channel.index,autopct="%1.1f%%",startangle=100,colors=plt.cm.Paired.colors)
plt.title("Distribution of total roi by channel used")
plt.tight_layout()
plt.show()
```

```python
####Scatter plot
plt.figure(figsize=(8,5))
sns.scatterplot(data=df,x="Impressions",y="Clicks",hue="Channel_Used",s=100)
plt.title("Relationships b/w impressions,clicksby channelused")
plt.xlabel("Impressions")
plt.ylabel("Clicks")
plt.legend(title="Channel_Used")
plt.tight_layout()
plt.show
```

```python
### Time-based analysis
### Histogram
plt.figure(figsize=(8,5))
plt.hist(df["Duration"],bins=30,edgecolor="black",color="red")
plt.title("Distribution of duration")
plt.xlabel("Duration")
plt.ylabel("frequency")
plt.show()
```

```python
### Line chart
df["Date"] = pd.to_datetime(df["Date"], format='%d-%m-%Y')
plt.figure(figsize=(8,5))
sns.lineplot(data=df,x="Date",y="Conversion_Rate",color="blue")
plt.title("Conversion rate over time")
plt.xlabel("Date")
plt.ylabel("Conversion_Rate")
plt.legend(title="company")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
### Line chart
df["Date"] = pd.to_datetime(df["Date"], format='%d-%m-%Y')
plt.figure(figsize=(8,5))
sns.lineplot(data=df,x="Date",y="Engagement_Score",color="blue",marker="o")
plt.title("Engagement_Scoren  over time")
plt.xlabel("Date")
plt.ylabel("Engagement_Score")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#### Bar chart
location_cost =df.groupby("Location")["Acquisition_Cost"].mean().reset_index()
location_cost = location_cost.sort_values(by="Acquisition_Cost",ascending=False)
plt.figure(figsize=(8,5))
sns.barplot(data=location_cost,x="Location",y="Acquisition_Cost",palette="viridis")
plt.title("Acquisition_Cost by  Location")
plt.xlabel("Location")
plt.ylabel("Acquisition_Cost")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
### Bar chart
df["Location"]=df["Location"].astype(str)
df["Target_Audience"]=df["Target_Audience"].astype(str)
plt.figure(figsize=(8,5))
sns.barplot(data=df,x="Location",y="Conversion_Rate",hue="Target_Audience",palette="Set2")
plt.title("conversion rate by location and target audience")
plt.xlabel("Location")
plt.ylabel("Conversion_Rate")
plt.xticks(rotation=45)
plt.legend(title="Target_Audience")
plt.tight_layout()
plt.show()
```

```python
###Pie chart
roi_by_location=df.groupby("Location")["ROI"].sum().reset_index()
plt.figure(figsize=(8,5))
plt.pie(
    roi_by_location["ROI"],
        labels=roi_by_location["Location"],autopct="%1.1f%%",startangle=140,wedgeprops={"edgecolor":"black"})
plt.title("Proportion of ROI by Location ")
plt.tight_layout()
plt.show()
```
